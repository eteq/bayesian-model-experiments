```python
import numpy as np

from scipy import stats
```

```python
%matplotlib inline
from matplotlib import rcParams, style
from matplotlib import pyplot as plt

rcParams['image.cmap'] = 'viridis'
rcParams['image.origin'] = 'lower'
rcParams['axes.prop_cycle'] = style.library['seaborn-deep']['axes.prop_cycle']
rcParams['figure.figsize'] = (14, 8)
rcParams['axes.titlesize'] =  rcParams['axes.labelsize'] = 16
rcParams['xtick.labelsize'] =  rcParams['ytick.labelsize'] = 14
```

$a \pm \sigma_a$

$b \pm \sigma_b$

are a and b consistent?

```python
a = 1
sa = 0.3

b = 1.
sb = 0.2

n = 10000

rva = sa*np.random.randn(n) + a
rvb = sb*np.random.randn(n) + b
```

```python
bs = np.linspace(min(np.min(rva), np.min(rvb)),
                   max(np.max(rva), np.max(rvb)),
                   100)

plt.hist(rva, histtype='step',bins=bs)
plt.hist(rvb, histtype='step', bins=bs)
plt.xlim(bs[0], bs[-1])
```

```python
a0,b0=rva[0], rvb[0]

mus = np.linspace(a0, b0, 1000)
prod = stats.norm.pdf(b0, mus, sb)*stats.norm.pdf(a0, mus, sa)
plt.plot(mus, prod)
plt.axvline(a0, c='k')
plt.axvline(b0, c='k')
```

```python

mi = min(a, b)
mx = max(a, b)
mus = np.linspace(mi-(sa+sb)*2, mx+(sa+sb)*2, 1000)

for i in range(10):
    ai, bi = rva[i], rvb[i]
    
    
    
    prod = stats.norm.pdf(bi, mus, sb)*stats.norm.pdf(ai, mus, sa)
    plt.plot(mus, prod)
plt.axvline(a, c='k')
plt.axvline(a-sa, c='k', ls=':')
plt.axvline(a+sa, c='k', ls=':')
plt.axvline(b, c='k')
plt.axvline(b+sb, c='k', ls=':')
plt.axvline(b-sb, c='k', ls=':')
```
